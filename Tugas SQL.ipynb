```python
# 1. Prepare a list of offices sorted by country, state, city
run("SELECT * FROM offices ORDER BY country, state, city")
```

```python
# 2. How many employees are there in the company?
run("SELECT COUNT(*) AS Jumlah_Pegawai FROM employees ")
```

```python
# 3.  What is the total of payments received?
run("SELECT SUM(amount) AS Total_Payments FROM payments ")
```

```python
# 4. List the product lines that contain 'Cars'.

run("""
SELECT 
    *
FROM
    productlines
WHERE
    productLine LIKE '%Cars%'
""")
```

```python
# 5. Report total payments for October 28, 2004.
run("""
SELECT SUM(amount) AS 'Payment 28okt04'
FROM payments 
WHERE paymentDate='2004-10-28'
""")
```

```python
# 6. Report those payments greater than \$100,000.
run("""
SELECT
    *
FROM
    payments
WHERE
    amount>100000
""")
```

```python
# 7. List the products in each product line.
run("SELECT productLine,productName FROM products ORDER BY productLine")
```

```python
# 8. How many products in each product line?
run("SELECT productLine,COUNT(*) AS JumlahProduk FROM products GROUP BY productLine")
```

```python
# 9. What is the minimum payment received?
run("SELECT MIN(amount) as Payment_Minimum FROM payments")
```

```python
# 10. List all payments greater than twice the average payment.
run("""
SELECT * 
FROM payments
WHERE amount > (SELECT AVG(amount)*2 FROM Payments)
""")
```

```python
# 11. What is the average percentage markup of the MSRP on buyPrice?
run("SELECT AVG((MSRP-buyPrice)/buyPrice)*100 AS average_markup  FROM Products")
```

```python
#12.  How many distinct products does ClassicModels sell?
run("SELECT COUNT(productCode) AS totalProducts FROM Products")
# SELECT COUNT(DISTINCT productCode) FROM Products
```

```python
# 13. Report the name and city of customers who don't have sales representatives?
run("SELECT customerName,city FROM customers WHERE customerNumber NOT IN (SELECT customerNumber FROM orders)")
```

```python
# 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("""
SELECT CONCAT(firstName,' ',lastName) AS fullName, jobTitle 
FROM employees 
WHERE jobTitle LIKE 'VP%' OR jobTitle LIKE '%Manager%'
""")
```

```python
# 15. Which orders have a value greater than $5,000?
run("""
SELECT O.orderNumber, P.amount
FROM orders O, customers C, payments P
WHERE C.customerNumber=O.customerNumber AND C.customerNumber=P.customerNumber AND P.amount>5000
""")
```
